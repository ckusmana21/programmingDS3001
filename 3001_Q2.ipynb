```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 50)
log_y = np.log(x)
exp_y = np.exp(x)

plt.scatter(x, log_y, label='Natural Log')
plt.scatter(x, exp_y, label='Exponential')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Natural Log and Exponential Functions')
plt.legend(loc='lower right')

plt.show()

x_grid = np.linspace(-6.5, 6.5, 100)
sin_y = np.sin(x_grid)
cos_y = np.cos(x_grid)

plt.plot(x_grid, sin_y, label='Sine')
plt.plot(x_grid, cos_y, label='Cosine')

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Sine and Cosine Functions')
plt.legend(loc='lower left')

plt.show()
```
